```python
# âœ… Impor pustaka yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# ðŸ“Œ Data dummy: GDP per kapita (USD) dan kepuasan hidup (skala 0â€“10)
gdp = np.array([5000, 8000, 10000, 12000, 15000, 18000, 20000, 23000, 25000, 30000,
               32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000]).reshape(-1, 1)
life_satisfaction = np.array([3.2, 3.5, 3.8, 4.0, 4.3, 4.6, 4.9, 5.1, 5.3, 5.6,
                              5.7, 5.9, 6.0, 6.2, 6.3, 6.5, 6.6, 6.8]).reshape(-1, 1)
```

```python
# ðŸ“Š Visualisasi awal hubungan GDP dan kepuasan hidup
plt.figure(figsize=(8,5))
plt.scatter(gdp, life_satisfaction, color='blue')
plt.title('Hubungan GDP per Kapita dan Kepuasan Hidup')
plt.xlabel('GDP per kapita (USD)')
plt.ylabel('Kepuasan hidup (0â€“10)')
plt.grid(True)
plt.show()
```

```python
# âš™ï¸ Buat dan latih model regresi linear
model = LinearRegression()
model.fit(gdp, life_satisfaction)
```

```python
# ðŸ“ˆ Informasi garis regresi
print("Kemiringan (slope):", model.coef_[0][0])
print("Intercept (titik potong Y):", model.intercept_[0])
```

```python
# ðŸ” Prediksi untuk beberapa nilai GDP baru
gdp_baru = np.array([[11000], [26000], [37000], [48000]])
prediksi = model.predict(gdp_baru)

for i, gdp_val in enumerate(gdp_baru.flatten()):
    print(f"GDP ${gdp_val:,} â†’ Prediksi kepuasan hidup: {prediksi[i][0]:.2f}")
```

```python
# ðŸ“ Evaluasi performa model
y_pred = model.predict(gdp)
mse = mean_squared_error(life_satisfaction, y_pred)
r2 = r2_score(life_satisfaction, y_pred)
print("\nMean Squared Error (MSE):", round(mse, 4))
print("Koefisien Determinasi (RÂ²):", round(r2, 4))
```

```python
# ðŸ–¼ Visualisasi hasil regresi
plt.figure(figsize=(8,5))
plt.scatter(gdp, life_satisfaction, label='Data Aktual', color='blue')
plt.plot(gdp, y_pred, label='Garis Regresi', color='red')
plt.scatter(gdp_baru, prediksi, color='green', marker='x', label='Prediksi Baru')
plt.title('Regresi Linear: GDP vs Kepuasan Hidup')
plt.xlabel('GDP per kapita (USD)')
plt.ylabel('Kepuasan hidup')
plt.legend()
plt.grid(True)
plt.show()
```
